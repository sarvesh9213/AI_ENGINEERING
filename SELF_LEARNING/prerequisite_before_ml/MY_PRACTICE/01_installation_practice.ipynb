```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib as plt 
from sklearn import linear_model

```

```python
# Create a Linear Regression model
reg = linear_model.LinearRegression()

# X contains 3 samples and 2 features
X = [[0, 0], [1, 1], [2, 2]]

# y is the target value for each sample
y = [0, 1, 2]

# Train the model to learn the relationship between X and y
reg.fit(X, y)

# Coefficient/weight learned for each input feature
print(reg.coef_)

# Constant term learned by the model
print(reg.intercept_)

# Predict target for a new sample
print(reg.predict([[3, 3]]))
```
