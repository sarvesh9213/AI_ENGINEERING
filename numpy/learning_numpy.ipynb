2. //////////////////// NumPy Arrays Basics///////////////// 

Arrays are the core of NumPy. 
Creating 1D, 2D, and 3D arrays 

1D array: like a list (row of values) 
2D array: like a table or spreadsheet
3D array: like a stack of spreadsheets 

///////////////Real-life Example////////////// 

1D: A row of student marks 
2D: A table of marks for 3 students in 3 subjects 
3D: Marks of 3 classes, each having a table of student-subject scores
