```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("D:\\desktop\\ML\\DATA\\50_Startups.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset['State'].value_counts()
```

```python
dataset = pd.get_dummies(dataset, columns=['State'])
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['State_California'], axis=1)
```

```python
dataset.head()
```

```python
# split the data into indep and dep
x = dataset.drop(['Profit'], axis=1)
y = dataset[['Profit']]
```

```python
x.head()
```

```python
y.head()
```

```python
# pls handle outlier if you compare with linear regression
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

# Building Model - DecisionTree Regression Model

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor()
dtree.fit(x_train, y_train)
```

```python
y_pred_train = dtree.predict(x_train)
y_pred_test = dtree.predict(x_test)
```

```python
# Evaluate the model
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_train, y_pred_train))
print()
print(r2_score(y_test, y_pred_test))
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train, y_train)
```

```python
y_pred_train = linear.predict(x_train)
y_pred_test = linear.predict(x_test)
```

```python
print(r2_score(y_train, y_pred_train))
print()
print(r2_score(y_test, y_pred_test))
```

# Random Forest Regression 

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
y_pred_train = rf.predict(x_train)
y_pred_test = rf.predict(x_test)
```

```python
print(r2_score(y_train, y_pred_train))
print()
print(r2_score(y_test, y_pred_test))
```
